```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
% matplotlib inline

wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["Teamcenter", "Blake", 'will', 'Mike', 'Pullen', 'Suchi', 'Andres', 'Chris', 'Villegas'])
```

```python
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="black", max_words=50, contour_width=2).generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (20, 20), facecolor = None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image

wine_mask = np.array(Image.open("M-Letter-PNG-High-Quality-Image.png"))
wine_mask.max()
print(wine_mask)
```

```python
def transform_format(val):
    if val == 26:
        return 255
    else:
        return val

# Transform your mask into a new one that will work with the function:
transformed_wine_mask = np.ndarray((wine_mask.shape[0],wine_mask.shape[1]), np.int32)

for i in range(len(wine_mask)):
    transformed_wine_mask[i] = list(map(transform_format, wine_mask[i]))
    
```

```python
transformed_wine_mask
```

```python
# Create a word cloud image
wc = WordCloud(background_color="white", max_words=1000, mask=transformed_wine_mask,
               stopwords=stopwords, contour_width=3, contour_color='firebrick')

# Generate a wordcloud
wc.generate(text)

# store to file
wc.to_file("wine.png")

# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
